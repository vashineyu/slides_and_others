```python
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
target_url = "https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt"
```

```python
df = pd.read_csv(target_url, sep="\t", header=None)
df.head()
```

```python
def img2arr_fromURLs(url_list, resize = False):
    img_list = []
    for url in url_list:
        response = requests.get(url)
        try:
            img = Image.open(BytesIO(response.content))
            if resize:
                img = img.resize((256,256)) # 假如 resize, 就先統一到 256 x 256
            img = np.array(img)
            img_list.append(img)
        except:
            # 只有在 response.status_code 為 200 時，才可以取得圖片，若有 404 或其他 status code, 會碰到 Error, 所以我們用 Try 語法避開取不到的狀況
            pass
    
    return img_list
```

```python
result = img2arr_fromURLs(df[0:5][1].values)
print("Total images that we got: %i " % len(result)) # 如果不等於 5, 代表有些連結失效囉

for im_get in result:
    plt.imshow(im_get)
    plt.show()
```

```python
def img2arr_fromURLs(url_list, resize = False, num_get=10):
    img_list = []
    counter = 0
    while len(img_list) < num_get:
        url = url_list[counter]
        response = requests.get(url)
        try:
            img = Image.open(BytesIO(response.content))
            if resize:
                img = img.resize((256,256)) # 假如 resize, 就先統一到 256 x 256
            img = np.array(img)
            img_list.append(img)
            print("URL: %s get, current get images: %i" % (url, len(img_list)))
        except:
            # 只有在 response.status_code 為 200 時，才可以取得圖片，若有 404 或其他 status code, 會碰到 Error, 所以我們用 Try 語法避開取不到的狀況
            print("URL: %s failed" % (url))
        counter += 1
    
    return img_list
```

```python
result = img2arr_fromURLs(df[1].values, num_get=5)
print("Total images that we got: %i " % len(result)) # 如果不等於 5, 代表有些連結失效囉

for im_get in result:
    plt.imshow(im_get)
    plt.show()
```
