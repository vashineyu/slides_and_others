```python
import pandas as pd
import numpy as np
import os
```

```python
fdir = "./mrt_data/"

csv_files = [os.path.join(fdir, i) for i in os.listdir(fdir)]
print(csv_files)
```

```python
pd.read_csv(csv_files[0], engine='python') # If you met UnicodeDecodeError, try encoding='utf-8 pr engine='python'
```

```python
df_list = [pd.read_csv(i, engine='python') for i in csv_files]
df = pd.concat(df_list)
df.reset_index(drop=True, inplace=True)
```
